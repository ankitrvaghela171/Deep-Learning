## Program 1: Tensor Creation and Basic Operations 
## Aim 
## To understand tensor creation and perform basic arithmetic operations using TensorFlow. 
## Algorithm 
### 1. Import TensorFlow library 
### 2. Create scalar, vector, and matrix tensors 
### 3. Perform addition, multiplication, and matrix multiplication 
### 4. Display results

```python


import tensorflow as tf

# Scalar (0-D Tensor)
scalar = tf.constant(10)
print("Scalar:", scalar)

# Vector (1-D Tensor)
vector = tf.constant([1, 2, 3])
print("Vector:", vector)

# Matrix (2-D Tensor)
matrix = tf.constant([[1, 2],
                       [3, 4]])
print("Matrix:\n", matrix)

# Addition
add_result = vector + 5
print("Addition:", add_result)

# Multiplication
mul_result = vector * 2
print("Multiplication:", mul_result)

# Matrix Multiplication
matmul_result = tf.matmul(matrix, matrix)
print("Matrix Multiplication:\n", matmul_result)




```

## Program 2: Linear Regression Using TensorFlow 
## Aim 
## To implement a simple linear regression model using TensorFlow. 
## Algorithm 
### 1. Import required libraries 
### 2. Create input and output data 
### 3. Define a linear model 
### 4. Compile the model with loss and optimizer 
### 5. Train the model 


```python

import numpy as np
import matplotlib.pyplot as plt
```

```python


import tensorflow as tf
import numpy as np

# Input and Output data
X = np.array([1, 2, 3, 4, 5], dtype=float)
y = np.array([2, 4, 6, 8, 10], dtype=float)

# Create Linear Regression Model
model = tf.keras.Sequential([
    tf.keras.layers.Dense(1, input_shape=(1,))
])

# Compile the model
model.compile(
    optimizer='sgd',
    loss='mean_squared_error'
)


# Train the model
model.fit(X, y, epochs=500, verbose=0)

# Test the model
prediction = model.predict(np.array([[8]]))
print("Prediction for x = 8:", prediction)


```

## Program 3: Simple Neural Network for Classification 
## Aim 
## To build and train a neural network for binary classification. 
## Algorithm 
### 1. Import TensorFlow and dataset 
### 2. Create a sequential neural network 
### 3. Add dense layers with activation functions 
### 4. Compile the model 
### 5. Train and evaluate the model

```python


import tensorflow as tf
from sklearn.datasets import make_classification
from sklearn.model_selection import train_test_split

# STEP 1: Create dataset (IMPORTANT FIX)
X, y = make_classification(
    n_samples=1000,
    n_features=2,
    n_informative=2,
    n_redundant=0,
    n_repeated=0,
    n_classes=2,
    random_state=42
)

# STEP 2: Split dataset
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.2
)

# STEP 3: Create Neural Network model
model = tf.keras.Sequential([
    tf.keras.layers.Dense(8, activation='relu', input_shape=(2,)),
    tf.keras.layers.Dense(1, activation='sigmoid')
])

# STEP 4: Compile model
model.compile(
    optimizer='adam',
    loss='binary_crossentropy',
    metrics=['accuracy']
)

# STEP 5: Train model
history = model.fit(X_train, y_train, epochs=20, verbose=1)

# STEP 6: Evaluate model
loss, accuracy = model.evaluate(X_test, y_test)
print("Accuracy:", accuracy)

# ---------------- GRAPH 1: Loss Graph ----------------
plt.figure()
plt.plot(history.history['loss'])
plt.xlabel("Epochs")
plt.ylabel("Loss")
plt.title("Training Loss Graph")
plt.show()

# ---------------- GRAPH 2: Accuracy Graph ----------------
plt.figure()
plt.plot(history.history['accuracy'])
plt.xlabel("Epochs")
plt.ylabel("Accuracy")
plt.title("Training Accuracy Graph")
plt.show()
```

## 4. Replace optimizer adam with sgd and analyze performance 

```python
# Exercise 4: Binary Classification using SGD Optimizer

import tensorflow as tf
from sklearn.datasets import make_classification
from sklearn.model_selection import train_test_split

# STEP 1: Create dataset
X, y = make_classification(
    n_samples=1000,
    n_features=2,
    n_informative=2,
    n_redundant=0,
    n_classes=2,
    random_state=42
)

# STEP 2: Split dataset
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.2
)

# STEP 3: Create Neural Network Model
model = tf.keras.Sequential([
    tf.keras.layers.Dense(8, activation='relu', input_shape=(2,)),
    tf.keras.layers.Dense(1, activation='sigmoid')
])

# STEP 4: Compile model with SGD optimizer (MAIN CHANGE)
model.compile(
    optimizer=tf.keras.optimizers.SGD(learning_rate=0.01),
    loss='binary_crossentropy',
    metrics=['accuracy']
)

# STEP 5: Train model
history = model.fit(X_train, y_train, epochs=20, verbose=1)

# STEP 6: Evaluate model
loss, accuracy = model.evaluate(X_test, y_test)
print("Accuracy with SGD:", accuracy)

# ---------------- GRAPH 1: Loss Graph ----------------
plt.figure()
plt.plot(history.history['loss'])
plt.xlabel("Epochs")
plt.ylabel("Loss")
plt.title("Training Loss Graph")
plt.show()

# ---------------- GRAPH 2: Accuracy Graph ----------------
plt.figure()
plt.plot(history.history['accuracy'])
plt.xlabel("Epochs")
plt.ylabel("Accuracy")
plt.title("Training Accuracy Graph")
plt.show()
```

## 5. Implement multi-class classification using Softmax activation

```python


import tensorflow as tf
from sklearn.datasets import make_classification
from sklearn.model_selection import train_test_split
from tensorflow.keras.utils import to_categorical

# Create multi-class dataset
X, y = make_classification(
    n_samples=1000,
    n_features=4,
    n_informative=4,
    n_redundant=0,
    n_classes=3,
    random_state=42
)

# Convert labels to categorical
y = to_categorical(y)

# Split dataset
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.2
)

# Create model
model = tf.keras.Sequential([
    tf.keras.layers.Dense(16, activation='relu', input_shape=(4,)),
    tf.keras.layers.Dense(3, activation='softmax')
])

# Compile model
model.compile(
    optimizer='adam',
    loss='categorical_crossentropy',
    metrics=['accuracy']
)

# Train model
history = model.fit(X_train, y_train, epochs=20, verbose=1)

# Evaluate model
loss, accuracy = model.evaluate(X_test, y_test)
print("Accuracy:", accuracy)
# ---------------- GRAPH 1: Loss Graph ----------------
plt.figure()
plt.plot(history.history['loss'])
plt.xlabel("Epochs")
plt.ylabel("Loss")
plt.title("Training Loss Graph")
plt.show()

# ---------------- GRAPH 2: Accuracy Graph ----------------
plt.figure()
plt.plot(history.history['accuracy'])
plt.xlabel("Epochs")
plt.ylabel("Accuracy")
plt.title("Training Accuracy Graph")
plt.show()
```

# Exercise Questions

## 1. What is a tensor?

### tensor is a multi-dimensional data structure used to store data in Deep Learning.
### 0-D tensor → Scalar (single value)
### 1-D tensor → Vector
### 2-D tensor → Matrix
### 3-D or more → Higher dimensional tensor
### Tensor is the basic data unit in TensorFlow.

## 2. Difference between NumPy arrays and TensorFlow tensors

### | NumPy Array                  | TensorFlow Tensor                  |
### | ---------------------------- | ---------------------------------- |
### | Works only on CPU            | Works on CPU and GPU               |
### | Used for numerical computing | Used for Deep Learning             |
### | No automatic gradient        | Supports automatic differentiation |
### | Static operations            | Optimized & scalable               |


## 3. What is an optimizer?

### An optimizer is an algorithm that updates the weights of a neural network to reduce the loss.
### Examples:
### SGD (Stochastic Gradient Descent)
### Adam
### RMSprop

### Optimizer helps the model learn better and faster.

## 4. Explain loss functions

### A loss function measures how wrong the model’s prediction is.
### Small loss → Good prediction
### Large loss → Poor prediction
### Examples:
### Mean Squared Error (Regression)
### Binary Crossentropy (Binary classification)
### Categorical Crossentropy (Multi-class classification)
### The goal of training is to minimize loss.
