### Exercise 1: Scalars and Vectors

```python
import numpy as np
bias =0.5
x = np.array([2.0, 3.0])
w = np.array([0.4, -0.6])
print("Bias (scalar):", bias)
print("Input vector x:", x)
print("Weight vector w:", w)
```

### Exercise 2: Dot Product

```python
dot_product = np.dot(w, x)
print("Dot product (w · x):", dot_product)
```

## Interpretation:
### • A positive value indicates the input lies on one side of the decision boundary
### • A negative value indicates the opposite side

### Exercise 3: Linear Combination

```python
z = dot_product + bias
print("Linear output (z = w·x + b):", z)
```

### Exercise 4: Step Activation Function


```python
def step_activation(z):
    if z >= 0:
        return 1
    else:
        return 0
output = step_activation(z)
print("Perceptron output after step activation:", output)
```

## Discussion Point:
### • Why does a perceptron use an activation function?
### A perceptron uses an activation function to Introduce non-linearity, Convert continuous values into class labelEnable decision-making (0 or 1)

### Exercise 5: Matrix as Batch of Samples

```python
X = np.array([
    [2.0, 3.0],
    [1.0, -1.0],
    [0.0, 2.0],
    [-1.0, -2.0]
])

print("Input matrix (batch of samples):")
print(X)
```

## Key Idea:
### • Deep learning models process data in batches for efficiency

### Exercise 6: Batch Dot Product

```python
# Batch dot product
Z = np.dot(X, w) + bias
print("Linear output for batch:")
print(Z)
```

## Insight:
### • This operation is identical to computations performed in deep neural networks
